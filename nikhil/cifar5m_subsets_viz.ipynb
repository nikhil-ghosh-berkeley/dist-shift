```python
%matplotlib inline
%reload_ext autoreload
import sys; sys.path.append('..')
import matplotlib.pyplot as plt
import numpy as np
import os
from src.simple_utils import load_pickle
from src.data_utils import cifar10_label_names, get_dataset
from os.path import join
```

```python
names = {0: "CIFAR10-neg", 1: "CIFAR10-neg-bal", 2: "CIFAR10-easy", 3: "CIFAR10-hard"}
name = names[0]
print("loading dataset")
dataset = get_dataset('../data', name)
data_x = dataset.data_x
data_y = dataset.data_y
class_names = cifar10_label_names
print("done")
```

```python
ncols = 5
nrows = 4
n = ncols * nrows

for c in range(len(class_names)):
    class_idx = np.where(data_y == c)[0]
    if len(class_idx) < n:
        continue
    class_idx_sub = class_idx[np.random.choice(len(class_idx), size=n, replace=False)]

    fig, axs = plt.subplots(nrows, ncols, figsize=(12, 9))
    for i in range(nrows):
        for j in range(ncols):
            img = data_x[class_idx_sub[i * ncols + j]]
            axs[i, j].imshow(img)
            axs[i, j].axis('off')

    fig.suptitle(class_names[c], fontsize=32)
    plt.show()
```
