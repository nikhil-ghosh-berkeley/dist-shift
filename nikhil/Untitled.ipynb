```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import interpolate
from scipy.ndimage import gaussian_filter1d
import numpy as np
```

```python
x = np.arange(0, 10)
y1 = np.exp(-x/3.0)
y2 = np.exp(-x/2.5)
Y = np.array([y1, y2])
f = interpolate.interp1d(x, Y)
```

```python
xnew = np.arange(0, 9, 0.1)
Ynew = f(xnew)
```

```python
Y_smooth = gaussian_filter1d(Ynew, sigma=2.0)
y1_new = Y_smooth[0]
y2_new = Y_smooth[1]
print(Y_smooth.shape)
```

```python
plt.plot(x, y1, 'o', xnew, y1_new, '-')
plt.plot(x, y2, 'o', xnew, y2_new, '-')
plt.show()
```
